```python
from torchkeras.ocr import Pipeline,plots,set_logging 
set_logging(verbose=True)
pipe = Pipeline()
```

```python
img_path = 'https://img2.baidu.com/it/u=2823416918,2748212696&fm=253&fmt=auto&app=138&f=JPEG?w=667&h=500'
results = pipe.ocr(img_path, det=True, rec=True, cls=True)
results 
```
